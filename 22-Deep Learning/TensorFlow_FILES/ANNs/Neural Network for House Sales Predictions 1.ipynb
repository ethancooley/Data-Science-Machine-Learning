```python
df.drop('id', axis=1, inplace=True)
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
# we don't fit to our test set b/c we don't want to assume prior information
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model = Sequential()

"""Typically we base the number of neurons in our layers
on the number of features in our data"""

model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

# Output layer- price prediction
model.add(Dense(1))

# Configure the model for training
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x=X_train, y=y_train.values, 
          validation_data=(X_test, y_test.values), 
          batch_size=128, epochs=400)
```

```python
pd.DataFrame(model.history.history)
```

```python
""" This shows the differences in loss from backpropogation
and loss when comparing to validation data """

losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python

# If we see the validation loss go up as loss continue to go down,
# this shows us that we are overfitting. However right here, 
# we do not see any overfitting occurring
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test, predictions)
plt.plot(y_test,y_test,'r')
```
